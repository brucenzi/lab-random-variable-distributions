```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv("files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df
```

### 1. Get the numerical variables from our dataset.

```python
customer_df_numerical = customer_df.select_dtypes(include=np.number)
print(customer_df_numerical.columns)
```

### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in customer_df_numerical.columns:
    sns.histplot(customer_df_numerical[col])
    plt.show()
```

### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result

```python
#LOG
customer_df_numerical_log = np.log(customer_df_numerical)
```

```python
for col in customer_df_numerical_log.columns:
    sns.histplot(customer_df_numerical_log[col])
    plt.show()
```

```python
# Log-tranform skew most of the plots to the opposite side of th original dataset
```

```python
#SQUARE
customer_df_numerical_log = np.sqrt(customer_df_numerical)
```

```python
for col in customer_df_numerical_log.columns:
    sns.histplot(customer_df_numerical_log[col])
    plt.show()
```

```python
# Sqrt-tranform improves the distribution towards normalization. Total Claim Amount has the more significant transformation towards normal 
```

```python
#CUBE
customer_df_numerical_log = np.cbrt(customer_df_numerical)
```

```python
for col in customer_df_numerical_log.columns:
    sns.histplot(customer_df_numerical_log[col])
    plt.show()
```

```python
# Cbrt-tranform has the best result towards normalization. 
# Total Claim Amount has its distribution normalized 
```
